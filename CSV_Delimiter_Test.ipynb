####  <font color='lightgreen'>The 'c' engine does not support separators greater than 1 character when encoded in UTF-8.<br> Therefore, pandas is falling back to the 'python' engine to read the CSV file.<br> The warning message also suggests that we can avoid this warning by specifying engine='python'.<br><br>  Since this character: <b><font color='red' size='5'>`ᚵ`</font></b>is not a standard delimiter, pandas is treating it as a regular character and trying to read the file using the specified encoding (UTF-16) </font>

NOTE: <br>
ISO-8859-1 is a character encoding standard that is part of the ISO/IEC 8859 series of ASCII-based standard character encodings. It is also known as Latin-1 and is used throughout the Americas, Western Europe, Oceania, and much of Africa 1. It encodes what it refers to as “Latin alphabet no. 1”, consisting of 191 characters from the Latin script 1. `"latin1"` is a Python alias for ISO-8859-1 character encoding. In other words, `"latin1"` and `"ISO-8859-1"` are almost the same encoding. It encodes the upper range of ISO 8859-1.

## Dynamically find the correct encoding

```python
import chardet

# file_path = r"C:\Temp\Movies_Custom_Delimeter.csv"
# file_path = r"C:\Temp\Movies_Custom_Delimeter1.csv"
file_path = r"C:\Temp\Movies_Custom_Delimeter2.csv"

# Detect the encoding
with open(file_path, 'rb') as rawdata:
    result = chardet.detect(rawdata.read(10000))

# Print the encoding
print(result['encoding'])

# Now use the detected encoding to read the file
# df = pd.read_csv(file_path, delimiter='~@~', encoding=result['encoding'], engine='python')
# df = pd.read_csv(file_path, delimiter='@', encoding='latin1')
# df = pd.read_csv(file_path, delimiter='ᚵ', encoding=result['encoding'])
df = pd.read_csv(file_path, delimiter='ᚵ', encoding=result['encoding'],  engine='python')

# Print the dataframe
df
```
